```python
from sklearn.preprocessing import LabelEncoder

# Example: Encode 'Participation'
if 'O_score' in df.columns:
    participation_encoder = LabelEncoder()
    df['O_score'] = participation_encoder.fit_transform(df['O_score'])

# Encode target column
target_encoder = LabelEncoder()
df['Career'] = target_encoder.fit_transform(df['Career'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot to visualize outliers
sns.boxplot(data=df[['O_score', 'C_score']])
plt.title("Check for Outliers")
plt.show()
```

```python
# Histogram of C_score
sns.histplot(df['C_score'], kde=True)
plt.title('C_score Distribution')
plt.show()

# Count plot of target variable
sns.countplot(x='Career', data=df)
plt.title('Career Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
